```python
# type is a callable class.
# There are two ways of using type.
# 1. one parameter form get the class of an object
type(7) #--> int, 7 is an instance of int class
type(int) #--> type, int is an instance of type class. Most of the common python classes are instance of type.
type(UserDict) #--> abc.ABCMeta, some exception to the above
# 2. 3 parameter form creates a class dynamically, second parameter are parent classes
MyClass = type('MyClass', (str,),
               {
                   'x': 42,
                   'f': lambda self: self.x * 2
               })
```
